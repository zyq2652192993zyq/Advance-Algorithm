```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = []
for i in range(34):
    column_names.append(f'feature_{i + 1}')
column_names.append('flag')
```

```python
df = pd.read_csv('./ionosphere.data', names=column_names)
```

```python
df.head
```

```python
df.describe()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.ensemble import GradientBoostingClassifier
```

```python
n, m = df.shape
print(f'n = {n}, m = {m}')
X = df.iloc[:, 0:m - 1]
y = df.iloc[:, -1]
print(X.shape, y.shape)
```

```python
kf = KFold(n_splits=10)
scores = []
for train_index, test_index in kf.split(df):
    train_x, test_x, train_y, test_y = X.iloc[train_index], X.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    gbdt = GradientBoostingClassifier(
        max_depth=4, max_features=9, n_estimators=100
    )
    gbdt.fit(train_x, train_y)
    predict_y = gbdt.predict(test_x)
    score = accuracy_score(test_y, predict_y)
    print(f'score = {score}')
    print('=' * 100)
    scores.append(score)

score_mean = np.mean(scores)
print(f'mean score = {score_mean}')
```
